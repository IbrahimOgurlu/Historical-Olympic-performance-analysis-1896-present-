```python
import pandas as pd
import matplotlib.pyplot as plt 

df=pd.read_csv("athlete_events.csv")

df.head()
```

```python
madalyalilar =df[df["Medal"].notna()]
top_10_ulke= madalyalilar["NOC"].value_counts().head(10)
```

```python
plt.figure(figsize=(12,6))
top_10_ulke.plot(kind="bar", color="teal", edgecolor="black")
plt.title("Olimpiyat Tarihinde En Çok Madalya Kazanan 10 Ülke", fontsize=14)
plt.xlabel("Ülke kodu (NOC)", fontsize=12)
plt.ylabel("Toplam Madalya Sayısı", fontsize=12)
plt.xticks(rotation=45)
plt.show()
```
