```python
import pandas as pd
import matplotlib.pyplot as plt

df= pd.read_csv("athlete_events.csv")

df.head()
```

```python
#yıllara göre eşsiz sporccu sayıları gruplandırıldı
sporcu_sayisi=df.groupby("Year")["ID"].nunique()
```

```python
#line plot
plt.style.use("seaborn-v0_8-darkgrid")
plt.figure(figsize=(12,6))


plt.plot(sporcu_sayisi.index, sporcu_sayisi.values, marker="o", color="b", linestyle="--")
plt.title("Yıllara Göre Olimpiyatlara Katılan Sporcu Sayısı", fontsize=15)
plt.xlabel("Yıl", fontsize=12)
plt.ylabel("Sporcu sayısı", fontsize=12)

plt.show()
```
