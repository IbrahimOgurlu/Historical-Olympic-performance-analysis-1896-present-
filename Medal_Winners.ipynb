```python
import pandas as pd
import matplotlib.pyplot as plt

df= pd.read_csv("athlete_events.csv")

df.head()
```

```python
madalyalilar=df[df["Medal"].notna()]
```

```python
plt.figure(figsize=(10,6))
plt.hist(madalyalilar["Age"].dropna(), bins=20, color="coral", edgecolor="black")
plt.title("Madalya Kazanan Sporcuların Yaş Dağılımı", fontsize=14)
plt.xlabel("Yaş", fontsize=12)
plt.ylabel("Sporcu Sayısı", fontsize=12)
plt.axvline(madalyalilar["Age"].mean(), color="red", linestyle="dashed", linewidth=2, label="Ortalama yaş")
plt.legend()
plt.show()
```
