```python
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv("athlete_events.csv")

df.head()
```

```python
basketbol= df[df["Sport"]== "Basketball"]["Height"].dropna()
jimnastik=df[df["Sport"]=="Gymnastics"]["Height"].dropna()

plt.figure(figsize=(8,6))
plt.boxplot([basketbol,jimnastik], labels=["Basketball", "jimnastik"], patch_artist=True)
plt.title("Basketbol ve Jimnastik Branşlarında Boy Dağılımı ve Aykırı Değerler", fontsize=14)
plt.ylabel("Boy (cm)", fontsize=12)
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```
